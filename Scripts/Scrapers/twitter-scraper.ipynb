```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
#tickers
ticker = ["$AAPL OR 'Apple Stock' OR #AAPL", "$MSFT OR 'Microsoft Stock' OR #MSFT", "$GOOG OR 'Google Stock' OR #GOOG", "$NVDA OR 'Nvidia Stock' OR #NVDA", \
          "$FB OR 'Meta Stock' OR 'Facebook Stock' OR #FB", "$ASML OR 'ASML Stock' OR #ASML", "$AVGO OR 'Broadcom Stock' OR #AVGO", "$ADBE OR 'Adobe Stock' OR #ADBE", \
          "$CSCO OR 'Cisco Stock' OR #CSCO", "$QCOM OR 'Qualcomm Stock' OR #QCOM"]

# Creating list to append tweet data to
tweets_list = []

# look through the list of tickers
for c in range(len(ticker)):
# Using TwitterSearchScraper to scrape data and append tweets to list
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(ticker[c] + ' since:2021-01-01 until:2022-01-01 lang:en filter:has_engagement').get_items()):
        tweets_list.append([tweet.date, tweet.id, tweet.content, tweet.likeCount, tweet.retweetCount, tweet.user.username])
```

```python
# Creating a dataframe from the tweets list above
tweets_df = pd.DataFrame(tweets_list, columns=['Datetime', 'Tweet Id', 'Text','Likes','Retweets','Username'])

# Drop the duplicates to remove spam tweets and keep only one
tweets_df.drop_duplicates(subset='Text',keep='first')

# Fix the text
tweets_df['Text'].replace(r'\s+|\\n', ' ', regex=True, inplace=True) 
tweets_df
```

```python
# Export dataframe into a CSV
tweets_df.to_csv('scraped-tweets-v1.csv', index=False)
```
